Plotting

```python
df = all_cond.copy()
```

```python
# drop rows with NaN for plotting
df.dropna(axis="index", how="any",inplace=True)
```

```python
print(df["TSH_measured"].unique())
print(df["T3_measured"].unique())
print(df["TT4_measured"].unique())
print(df["T4U_measured"].unique())
print(df["FTI_measured"].unique())
```

```python
import plotly.express as px
# look for outliars in the numerical data
numerical_features = ["age", "TSH", "T3", "TT4", "T4U", "FTI"]
for feature in numerical_features:
  fig = px.box(df, y=feature)
  fig.show()

```

```python
# drop the age outlier
df = df[df["age"] <= 100]
df["age"].max()
```

```python
measurements = ["TSH", "T3", "TT4", "T4U", "FTI"]
for measure in measurements:
  fig = px.scatter(df, x="age", y=measure, title=f"age vs. {measure}")
  fig.show()
```

```python
other_attributes = ["sex", "target_hyper", "target_hypo"]
measurements = ["TSH", "T3", "TT4", "T4U", "FTI"]

for measure in measurements:
  fig = px.scatter(df, x="age", y=measure, color="sex", title=f"age vs. {measure}")
  fig.show()
```

```python
for measure in measurements:
  fig = px.scatter(df, x="age", y=measure, color="target_hyper", title=f"age vs. {measure}")
  fig.show()
```

```python
for measure in measurements:
  fig = px.scatter(df, x="age", y=measure, color="target_hypo", title=f"age vs. {measure}")
  fig.show()
```

Compare the datasets

```python
all_cond = pd.read_csv("thyroid-disease-data/all_conditions.csv")
all_thy = pd.read_csv("thyroid-disease-data/all_thyroid.csv")

# compare all_thy and all_cond

all_cond['patient_id'].equals(all_thy['patient_id'])

# patient ids are the same
```

```python
all_cond.info()
```

```python
all_thy.info()
```

```python
# Compare all_cond and all_thy
all_cond_columns = list(all_cond)
print(all_cond_columns)
all_cond_columns = all_cond_columns[:32]
print(all_cond_columns)

print(all_cond[all_cond_columns].equals(all_thy))

# The only difference between the files is all_cond has three more features
```

```python
# Compare all_cond_test and all_thy_test
all_cond_test = pd.read_csv("thyroid-disease-data/all_conditions_test.csv")
all_thy_test = pd.read_csv("thyroid-disease-data/all_thyroid_test.csv")
```

```python
# Compare all_cond test and all_thy test
print(all_cond_test[all_cond_columns].equals(all_thy_test))

# The test data in the two files is the same as well, except for the 3 additional columns
```

```python
# Compare the train and test data
print(all_cond.equals(all_cond_test))
print(all_thy.equals(all_thy_test))

```
